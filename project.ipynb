### Creating derived columns :
To gain additional insights from the existing data, we can create new columns.One useful metric in this dataset is the Discount Percentage, which shows how much discount each laptop offers compared to its original price (MRP).

The formula used is:

Discount Percentage=(MRP−Current Price)/MRP×100

```python
df['Discount_Percentage'] = ((df['MRP'] - df['Current_Price']) / df['MRP']) * 100
df['Discount_Percentage'] = df['Discount_Percentage'].round(2)
```

```python
df
```

### Filtering and Aggregating Data
#### 1. Filtering top expensive laptops
To better understand pricing patterns in the laptop market, we can sort and filter the dataset.
Here, we identify the Top 10 most expensive laptops listed on Flipkart based on their current selling price.

```python
# Sort laptops by Current_Price in descending order
price_desc = df.sort_values(by='Current_Price', ascending=False)

# Display top 10 most expensive laptops
print(price_desc[['ProductName', 'Current_Price']].head(10))
```

##### 2. Identifying 5-Star Rated Laptops

In this step, we focus on identifying laptops that have received a perfect 5-star rating from customers.
These products indicate maximum customer satisfaction, and analyzing them can help understand what features lead to the best reviews.

```python
top_rated = df[df['Stars'] == 5]
print(top_rated[['ProductName', 'Stars']])
```

##### 3. Brand Distribution Analysis

To understand which brands dominate the Flipkart laptop market, we extract the brand name from the ProductName column.
This is done by taking the first word of each product name, as it usually represents the brand (e.g., HP Pavilion, Lenovo IdeaPad).
Then, we count how many laptop models each brand offers

```python
# Create a new column called 'Brand' by taking the first word of the product name
df['Brand'] = df['ProductName'].str.split().str[0]

# Count how many laptops are there for each brand
count= df['Brand'].value_counts().reset_index()

# Show the top 5 brands
print(count.head(10))
```

##### 4. Brand-Wise Average Price and Rating
To understand how each brand performs in terms of pricing and customer satisfaction, we calculate the average Current Price and average Star rating for each brand.This gives insights into which brands sell premium laptops and which provide better-rated, budget-friendly options.

```python
brand_summary = df.groupby('Brand')[['Current_Price', 'Stars']].mean().round(2)
print(brand_summary)
```

# 4.visualisation
#### 1. Distribution of Laptop Prices

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Current_Price"], kde=True,color="green")
plt.title("Price Distribution of Laptops")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

The histogram shows that most laptops are priced between ₹30,000 and ₹60,000. The distribution is right-skewed, meaning a few high-end laptops raise the upper range, but premium models are less common. The KDE curve shows small peaks indicating budget and mid-range clusters. Overall, the laptop market is mainly affordable to mid-range.

#### 2. Distribution of Star Ratings

```python
plt.figure(figsize=(10,5))
df['Stars'].plot(kind='hist', bins=20, edgecolor='black')
plt.title('Distribution of Star Ratings')
plt.show()
```

This histogram shows how laptop ratings are distributed in the dataset. Most laptops have ratings between 3.5 and 4.5 stars, indicating generally positive customer feedback. A small number of entries with extremely low ratings (0 stars) may represent missing or invalid rating values. The plot helps understand overall product satisfaction levels.

#### 3. Top 10 most expensive Laptops (horizontal bar)

```python
plt.figure(figsize=(10,5))
top10 = df.nlargest(10, 'Current_Price')
plt.barh(top10['ProductName'], top10['Current_Price'],color="violet")
plt.gca().invert_yaxis()
plt.title('Top 10 Most Expensive Laptops')
plt.xlabel('Price (₹)')
plt.show()

```

This graph shows the top 10 most expensive laptops on Flipkart, ranked from the highest-priced at the top. It highlights premium models from ASUS, Lenovo, Dell, Acer, and Samsung, showcasing high-end gaming and productivity laptops with advanced specifications and brand value

#### 4.  Top 10 highest rated (by Stars)

```python
plt.figure(figsize=(10,5))
top_rated = df.nlargest(10, 'Stars')
plt.barh(top_rated['ProductName'], top_rated['Stars'],color="brown")
plt.gca().invert_yaxis()
plt.title('Top 10 Highest Rated Laptops')
plt.xlabel('Stars')
plt.show()
```

This graph shows the top 10 highest-rated laptops on Flipkart, ranked from the highest-rated at the top to lower-rated models below. It highlights customer preferences across brands like HP, ASUS, Acer, Lenovo, and Dell, showcasing the laptops that are most popular and highly appreciated by buyers.

Key insights: realme, Mi, and SAMSUNG offer the most expensive laptops on average, close to ₹60,000.AXL, Ultimus, and Primebook have the lowest average prices, below ₹25,000.Mid-range brands such as HP, DELL, ASUS, and acer fall between ₹45,000 – ₹50,000.

#### 7. Scatter Plot — Price vs Rating

#### 9. RAM Distribution (Count Plot)

#### 10. Heatmap — Correlation Matrix

```python
plt.figure(figsize=(8,5))
sns.heatmap(
    df[["Current_Price", "MRP", "Stars", "Ratings", "Reviews", "Discount_Percentage"]].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap of Laptop Features")
plt.show()
```

The heat map shows the correlation between key laptop features from Flipkart's dataset. Current price and MRP have a strong positive correlation( 0.83),indicating that higher-priced laptops generally have higher-listed prices. Ratings and reviews are also highly correlated( 0.98), meaning products with more reviews tend to have higher rating counts. Most other features show weak or negative correlations, such as discount percentage with current price(- 0.21), suggesting that cheaper laptops often receive higher discounts. Overall, the heat map highlights a few strong relationships, while most laptop features remain loosely connected.

#### 11. Pie Chart — Top 5 Brands Share

```python
top_brands = df["Brand"].value_counts().nlargest(5)
plt.figure(figsize=(6,6))
top_brands.plot(kind='pie', autopct='%1.1f%%')
plt.title("Top 5 Laptop Brands Share")
plt.ylabel("")
plt.show()
```

The pie chart displays the market share of the top five laptop brands based on Flipkart sales data. ASUS leads with the highest share of (25.2%), followed closely by HP (22.6%) and Dell (22.4%). Lenovo holds 19.9%, while Acer has the smallest share at 9.8%. The chart highlights a competitive market with Asus slightly ahead of the other major brands.

#### 12. . RAM vs Price (Bar Plot)

```python
plt.figure(figsize=(10,5))
df.groupby("Ram")["Current_Price"].mean().plot(kind="bar",color="black")
plt.title("Average Price by RAM Size")
plt.xlabel("RAM")
plt.ylabel("Average Price")
plt.show()
```

This graph depicts the average price of laptops on Flipkart based on their RAM size. It highlights how laptop prices generally vary with RAM capacity, showing trends such as higher prices for laptops with larger RAM (16 GB or 32 GB) compared to those with lower RAM (4 GB or 8 GB). # 5. Insight Generation and Report 
### 1. Summary of Key Insights


- **Price Distribution:**  
  Most laptops are priced between ₹30,000 and ₹60,000, indicating that Flipkart primarily focuses on the mid-range segment. Budget (<₹30,000) and premium (>₹60,000) laptops are less common.

- **Discount Patterns:**  
  Entry-level laptops (e.g., Lenovo IdeaPad 3) often have high discounts (~41%), while premium models (e.g., realme Book) have smaller discounts (~14%). This suggests targeted discounting to drive volume for lower-priced laptops.

- **Ratings and Reviews:**  
  Customer ratings (Stars) are generally high (4.0–4.5), regardless of price. The number of reviews varies widely, but there is no strong correlation between price and rating, suggesting mid-range laptops can achieve similar satisfaction as higher-priced models.

- **RAM and Storage Trends:**  
  4–8 GB RAM laptops are most common, with SSD storage (256–512 GB) being standard. This indicates the majority of buyers prioritize basic performance and speed over extreme specs.

- **Brand and Model Distribution:**  
  Lenovo, ASUS, and realme dominate the listings. Popular models include Lenovo IdeaPad 3, ASUS VivoBook 15, and realme Book. Established brands and widely used configurations are most visible on Flipkart.

- **Price vs Performance:**  
  Higher RAM and newer generation processors tend to increase prices moderately. Some models offer high performance at relatively lower prices, representing good value-for-money options for buyers.

- **Correlation Insights (Heat Map):**  
  - Strong positive correlation between Current_Price and MRP (~0.9)  
  - Moderate positive correlation between Current_Price and RAM (~0.5)  
  - Weak correlation between price and ratings or reviews  
  - Storage size has a slight positive correlation with price  

### 2. Patterns, Anomalies, and Correlations

- **Patterns:** Mid-range laptops dominate both listings and buyer ratings. Discounts are often strategically applied to drive volume on entry-level laptops.  
- **Anomalies:** Some high-priced laptops have fewer ratings, indicating either newer models or niche offerings.  
- **Correlations:** Price correlates strongly with MRP and moderately with RAM, indicating that performance features drive pricing. Customer satisfaction is largely independent of price.
